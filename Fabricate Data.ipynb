```python
my_list = []
counter = 1
for i in range(100):
    my_dict = {}
    my_dict['id'] = 'FAKEID' + str(counter)
    my_dict['first'] = names.get_first_name(gender='female')
    my_dict['last'] = names.get_last_name()
    my_dict['sex'] = 'female'
    #my_dict['age'] = round(random.gauss(30,10))
    my_dict['age'] = round(scipy.stats.truncnorm.rvs((18-35)/10,(60-35)/10,loc=35,scale=10,size=1)[0])
    my_dict['height'] = round(random.gauss(165.1,8.9))
    my_dict['weight'] = round(random.gauss(72.7,10),1)
    fake_add = real_random_address()
    for key, value in fake_add.items():
        my_dict[key] = value
    ph_no = []
    ph_no.append(random.randint(1, 9))
    for i in range(1, 10):
        ph_no.append(random.randint(0, 9))
    my_dict['phone'] = "".join(str(element) for element in ph_no)
    my_list.append(my_dict)
    counter += 1
for i in range(100):
    my_dict = {}
    my_dict['id'] = 'FAKEID' + str(counter)
    my_dict['first'] = names.get_first_name(gender='male')
    my_dict['last'] = names.get_last_name()
    my_dict['sex'] = 'male'
    #https://stackoverflow.com/questions/18441779/how-to-specify-upper-and-lower-limits-when-using-numpy-random-normal
    my_dict['age'] = round(scipy.stats.truncnorm.rvs((18-35)/10,(60-35)/10,loc=35,scale=10,size=1)[0])
    my_dict['height'] = round(random.gauss(177.8,10.16))
    my_dict['weight'] = round(random.gauss(86.4,10),1)
    fake_add = real_random_address()
    for key, value in fake_add.items():
        my_dict[key] = value
    ph_no = []
    ph_no.append(random.randint(1, 9))
    for i in range(1, 10):
        ph_no.append(random.randint(0, 9))
    my_dict['phone'] = "".join(str(element) for element in ph_no)
    my_list.append(my_dict)
    counter += 1
my_list
df = pd.DataFrame(my_list)
df
```

```python
# get the current time to name file
now = datetime.now()

# YYYY-mm-dd-H-M-S
dt_string = now.strftime("%Y%m%d%H%M%S")

# path for final report. We'll define an Excel writer object and the target file
with pd.ExcelWriter("C:/Users/cimacint/Downloads/" + dt_string + "_FabricatedData" + ".xlsx",engine="xlsxwriter") as writer:
    df.astype(str).to_excel(writer, sheet_name="FakeData",index=False)
    

print("Done.")
```
